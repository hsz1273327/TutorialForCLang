# 包管理

上文说道如何编译我们的源码为可执行文件,以及如何编译动态库,静态库.C/C++不像其他很多语言有成熟的包管理工具.不过也不是没有包管理工具,比较靠谱的一个是[conan](http://docs.conan.io/en/latest/index.html).它是一个C/C++的编译工具,集成了本地包管理,远程包下载,包发布和自动编译功能.不过毕竟不是官方维护的,里面收录包的极其不全,好处就是可以用统一的格式来管理包,体验上来说还行.


## 远程依赖管理

conan提供了一个完整的远程依赖下载和发布服务,并且可以通过命令将自己的远程服务和用户名密码注册到本地.

### 远端服务器

`conan`官方可以发布和下载C/C++的库,支持这个功能的服务器有俩:

+ `conan-transit`是个只读库,不会有更新,用于下载.
+ `conan-center`用于发布新的库.

而conan包提供一个简单的自建服务器方法--使用如下命令安装依赖:

```shell
git clone https://github.com/conan-io/conan.git
cd conan
git checkout master
pip install -r conans/requirements.txt
pip install -r conans/requirements_server.txt
pip install gunicorn
```
之后使用`gunicorn -b 0.0.0.0:9300 -w 4 -t 300 conans.server.server_launcher:app`来启动服务,而设置文件可以定义在`~/.conan_server/server.conf`之中,默认的配置如下:

```config
[server]
jwt_secret: MnpuzsExftskYGOMgaTYDKfw
jwt_expire_minutes: 120

ssl_enabled: False
port: 9300
public_port:
host_name: localhost

store_adapter: disk
authorize_timeout: 1800

# Just for disk storage adapter
disk_storage_path: ~/.conan_server/data
disk_authorize_timeout: 1800

updown_secret: NyiSWNWnwumTVpGpoANuyyhR


[write_permissions]
# "opencv/2.3.4@lasote/testing": default_user,default_user2

[read_permissions]
# opencv/1.2.3@lasote/testing: default_user default_user2
# By default all users can read all blocks
*/*@*/*: *

[users]
demo: demo
```

conan使用`conan search xxx -r conan-transit`来搜索远端库,功能类似`pip search`,如果不搜索包名,那就相当于`pip list
`.

要将远端的包安装到本地,可以使用形如`conan install OpenCV/3.2.0@xmar/stable -r conan-transit --build OpenCV`的形式安装.安装好后就在`~/.conan/data`文件夹下了.

### 本地依赖管理

conan下载下来的库会存放在`~/.conan/data`文件夹下.同样是使用`conan search xxx`来搜索本地缓存中的依赖.

### 管理远程服务器

conan有`remote`命令来将远程服务器注册到本地.添加使用`add`子命令,删除使用`remove`,查看注册的远程服务器有哪些可以使用`list`

同时可以使用`user`命令来将对应的用户注册给远端服务器

```shell
conan user [-h] [-p PASSWORD] [--remote REMOTE] [-c] [name]
```
